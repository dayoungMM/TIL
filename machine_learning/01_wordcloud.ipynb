```python
from nltk.tokenize import word_tokenize
from nltk import Text

# 앨리스 소설에에서 말뭉치 불러오기
carrol_raw = nltk.corpus.gutenberg.raw("carroll-alice.txt")


text = Text(word_tokenize(carrol_raw ))

```

```python
import matplotlib.pyplot as plt
text.plot(20) #가장 많이 사용하는 20개 단어의 빈도 나타내기
plt.show()
```

```python
fd = text.vocab()
type(fd)
```

```python
from nltk import FreqDist
from nltk.tokenize import RegexpTokenizer
from nltk.tag import pos_tag

retokenize = RegexpTokenizer("[\w]+")

stopwords = ["Mr.", "Mrs.", "Miss", "Mr", "Mrs", "Dear"] #이 단어들은 제외
emma_tokens = pos_tag(retokenize.tokenize(carrol_raw)) #단어 토큰화
names_list = [t[0] for t in emma_tokens if t[1] == "NNP" and t[0] not in stopwords] #stopwords에 없으면서 고유명사인 것 출력
fd_names = FreqDist(names_list) #자주나오는 단어 사용빈도 정보 출력
```

```python
fd_names
```

```python
from wordcloud import WordCloud

wc = WordCloud(width=1000, height=600, background_color="white", random_state=0)
plt.imshow(wc.generate_from_frequencies(fd_names))
plt.axis("off")
plt.show()
```
