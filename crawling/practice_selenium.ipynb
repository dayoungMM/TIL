```python
# 2. 특정영역의 엘리먼트 좌표값 구하기
driver = webdriver.Chrome()
driver.get(url)
```

```python
element = driver.find_element_by_css_selector('#movie_player')
location = element.location
size = element.size
print(location, size)
driver.quit()
```

```python
left = location["x"]*1.5
top = location["y"]*1.8
right = left + size["width"]*1.5
bottom = top + size["height"]*1.8
area = (left, top, right, bottom)
area
```

```python
# 3. 이미지를 pillow로 crop
pil_element_img = ipn_img.crop(area)
pil_element_img
```

```python
pil_element_img.save("datas/ipn_img_crop.png")
```
