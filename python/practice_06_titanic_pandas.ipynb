```python
# 필요한 컬럼을 골라서 새로운 데이터프레임 만들기
df_1= titanic_df[
    ["Survived","Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"   
    ]
]
df_1.tail()
```

#### 2. Nan 데이터 제거하기

```python
df_1.describe()
#age의 갯수가 다른 열들에 비해 작다는 것을 알 수 있다.
```

```python
# 방법 1 age가 비어있지 않은 행들만 골라서 나열
df_2=df_1[df_1["Age"].notnull()]
df_2.tail()
```

```python
len(df_2)
```

```python
df_3= df_1[df_1["Sex"].notnull()]
len(df_3)
```

```python
df_3= df_1[df_1["Embarked"].notnull()]
len(df_3)
```

```python
df_1.notnull().all(axis=0)
```

```python
df_1.notnull().all(axis=1)
```

```python
#어떤 컬럼에도  NAN값이 없는 행만 뽑아서 result로 만들기
result=df_1[df_1.notnull().all(axis=1)]
result.tail()
```

```python
#index 재정렬
result.reset_index(drop=True, inplace=True)
result.tail()
```

#### 3. One Hot Encoding

```python
#깊은복사
one_hot= result.copy()
```

```python
one_hot.tail()
```

```python
one_hot_1= pd.get_dummies(one_hot["Sex"])
one_hot_1.tail()
```

```python
one_hot_2 = pd.get_dummies(one_hot["Embarked"])
one_hot_2.tail()
```

```python
one_hot_result=pd.concat([one_hot,one_hot_1, one_hot_2],axis=1)
one_hot_result.tail()
```

```python
one_hot_result.notnull().all(axis=0)
```

```python
one_hot_result.drop(columns=["Sex","Embarked"], inplace=True)
```

```python
one_hot_result.tail()
```

#### 4. 연령대와 Adult 칼럼 만들기

```python
Final_df= one_hot_result.copy()
```

```python
Final_df.tail(3)
```

```python
Final_df["Ages"]=Final_df["Age"]//10*10
Final_df.tail()
```

```python
Final_df.drop(columns=["Age"], inplace=True)
Final_df.tail()
```

```python
Final_1=Final_df.copy()
```

```python
Final_1["Adult"]=0
Final_1.loc[Final_1["Ages"]>=20, "Adult"]=1
```
