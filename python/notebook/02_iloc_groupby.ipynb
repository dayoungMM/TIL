### 1962년 데이터에서 1962년 pop 평균 이하의 데이터에 대해 contry와pop ,gdpPercap열만 출력하라

```python
pop_mean=df[df.year==1962]['pop'].mean()
# idx=df[df.year==1962]['pop']<=pop_mean
# df[df.year==1962][idx][['pop','gdpPercap']]
```

```python
pop_mean=df[df['year']==1962]['pop'].mean()
df[(df['year']==1962) &(df['pop'] <= pop_mean)][['country','pop','gdpPercap']]
```

### 한국, 중국, 일본에 대한 데이터만을 뽑아서 year 부터 gdpPercap까지 표시하는데, year로 정렬하여 출력하라

## 그룹화한 데이터의 평균 구하기(44쪽)

### 6. lifeExp, gdpPercap 열의 평균값을 연도, 지역별로 그룹화하여 한 번에 계산하기

```python
multi_group_var = df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean() 
print(multi_group_var)
```

```python
multi_group_var = df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean() 
print(type(multi_group_var))
multi_group_var.head()
```

```python
multi_group_var.groupby(level=1)['gdpPercap'].mean()  #contiment 별로 groupby를 할 수 있음
```
