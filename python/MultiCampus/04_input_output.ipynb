### Pickle로 저장 및 읽기
- 문자열, 숫자인지 상관없이 바이너리 형식으로 읽고 저장
-  피클로 저장하는것은 가능

```python
# 파일 쓰기
import pickle
file = open('txt_files/list.bin', 'wb')
list1 = ['1',2,3,4]
pickle.dump(list1, file)
file.close
```

```python
file.close()
```

```python
# 파일 읽어오기
file = open('txt_files/list.bin', 'rb')
result = pickle.load(file)
result
```

### readline : 한줄씩 출력
- 끝까지 한줄씩 출력하고자 할 때에는 while 이용하기
- 텍스트 파일에 \n이 있는 경우를 조심하자

```python
# 끝까지 출력
file = open('txt_files/gugudan.txt') #rt 기본값
while True:
    data = file.readline()
    if not data:
        break
    print(data, end='') #파일 자체에 \n이 포함되어있어서 print 할 때 굳이 \n이 추가되지 않아도 된다
file.close()
```

```python
### 모든 내용을 리스트로 읽은 후 출력

file = open('txt_files/gugudan.txt') #rt 기본값
data = file.readlines() #리스트 형식으로 모두 가져온 후 
for d in data:  #출력하기 위해 loop
    print(d, end='')
file.close()
```

### with open
- close 불필요

```python
file = open('txt_files/gugudan.txt') #rt 기본값
data = file.readlines() #리스트 형식으로 모두 가져온 후 
with file:
    for d in data:
        print(d, end='')
```

```python
with open('txt_files/gugudan.txt') as file:
    data = file.readlines() #리스트 형식으로 모두 가져온 후 

for d in data:
    print(d, end='')
```

### Class 생성자 코드 with 로 불러오기

```python
class MyFile:
    def __init__(self):
        print('생성')
    def __enter__(self):      #class를 열때
        print('enter')   
    def __exit__(self, type, value, trackback):
        print('exit')   #class를 닫을때
        self.close()
    def close(self):
        pass
    
```
