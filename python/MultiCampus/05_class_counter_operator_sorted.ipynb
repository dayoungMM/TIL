```python
class Letter:
    letter = ''
    
    def write(self, text):
        self.letter += text
    
    def show(self):
        return self.letter
#   
letter1 = Letter()
#write   self 
letter1.write('')
letter1.write('')
letter2 = Letter()
letter2.write('ABCD')
letter2.write('WXYZ')
print(letter1.show())
print(letter2.show())
```

```python
class Book:
 #  
    title=''
    price = 0
    
    def setTitle(self, title):
        self.title = title
        return title

    def setPrice(self, price):
        self.price = price
        return price
    
book1 = Book()
book1.setTitle('Python')
book1.setPrice(10000)
book2 = Book()
book2.setTitle('Java')
book2.setPrice(10000)
print(book1.title, book1.price)
print(book2.title, book2.price)
```

```python
class Car:
 #  
    def __init__ (self):
        self.option_ls =[]
 # add_option  
    def add_option(self, option):
        self.option_ls.append(option)
 # show_option  
    def show_option(self):
        return self.option_ls

car1 = Car()
car2 = Car()
car1.add_option(' ')
car1.add_option(' ')
car2.add_option(' ')
car2.add_option('')
print(car1.show_option())
print(car2.show_option())
```

```python
class Pizza:
    def __init__(self):
        self.topping = []
    
    def add_topping(self, topping):
        self.topping += [topping]
    
    def remove(self, topping):
        self.topping.remove(topping)
        
    def show_topping(self):
        return self.topping
    
    
p1 = Pizza()
p1.add_topping('')
p1.add_topping('')
p1.add_topping('')
p2 = Pizza()
p2.add_topping('')
p2.add_topping('')
p2.add_topping('')
p2.remove('')
print(p1.show_topping())
print(p2.show_topping())
```

```python
class Product:
    def __init__ (self, name, code, count):
        self.name = name
        self.code = code
        self.count = count
        
    def show_info(self):
        return 'name=' + self.name + ', code=' + self.code + ', count=' + str(self.count)


p1 = Product('clock', 'c0001', 500)
print(p1.show_info())
p2 = Product('tv', 't0001', 1500)
print(p2.show_info())
```

 ###      EachSum  compute()  

```python
class EachSum:
    def __init__(self):
        self.total = 0

 #  
    def compute(self,num):
        result = sum([int(i) for i in str(num)])
        return result
each = EachSum()
total = each.compute(12345)
print('', total)
```

###      get_menu()     SelectMenu   (random  )

```python
import random
class SelectMenu:
    def __init__(self, menu_ls):
        self.menu_ls = menu_ls
        
    def get_menu(self):
        choice = random.choice(self.menu_ls)
        return choice
    
    def get_menu2(self):
        choice = self.menu_ls[random.randint(0,len(self.menu_ls)-1)]
        return choice
    
 #  
menu = SelectMenu(['', '', '', ''])
print(menu.get_menu())
print(menu.get_menu2())
```

```python
with open('txt_files/[ex]1-document.txt', 'rt') as file:
    text = file.read()
text = text.replace('\n', ' ')
words = text.split(' ')
word_dict = dict()

for word in words:
    if word in word_dict:
        word_dict[word] = word_dict[word] +1
    else:
        word_dict[word] = 1
print(word_dict)
    
```
