아래와 같은 실행 결과가 나올 수 있도록 Letter 클래스 코드 작성
 - write 메소드 - 편지내용 작성
 - show 메소드 – 작성된 내용 반환

```python
class Letter:
    letter = ''
    
    def write(self, text):
        self.letter += text
    
    def show(self):
        return self.letter
#생성된 인스턴스의 주소 저장
letter1 = Letter()
#write 호출시 주소가 self에 전달
letter1.write('가나다라')
letter1.write('카타파하')
letter2 = Letter()
letter2.write('ABCD')
letter2.write('WXYZ')
print(letter1.show())
print(letter2.show())
```

```python
class Book:
 # 코드 작성
    title=''
    price = 0
    
    def setTitle(self, title):
        self.title = title
        return title

    def setPrice(self, price):
        self.price = price
        return price
    
book1 = Book()
book1.setTitle('Python')
book1.setPrice(10000)
book2 = Book()
book2.setTitle('Java')
book2.setPrice(10000)
print(book1.title, book1.price)
print(book2.title, book2.price)
```

아래 코드가 실행되어 결과와 같이 나타날수 있도록 Car 클래스 작성
 - add_option() - 자동차 옵션 추가(List append)
 - show_option() - 추가된 옵션 목록 반환


```python
class Car:
 # 생성자 작성
    def __init__ (self):
        self.option_ls =[]
 # add_option 메소드 작성
    def add_option(self, option):
        self.option_ls.append(option)
 # show_option 메소드 작성
    def show_option(self):
        return self.option_ls

car1 = Car()
car2 = Car()
car1.add_option('전동 트렁크')
car1.add_option('통풍 시트')
car2.add_option('뒷자리 에어백')
car2.add_option('하이패스')
print(car1.show_option())
print(car2.show_option())
```

```python
class Pizza:
    def __init__(self):
        self.topping = []
    
    def add_topping(self, topping):
        self.topping += [topping]
    
    def remove(self, topping):
        self.topping.remove(topping)
        
    def show_topping(self):
        return self.topping
    
    
p1 = Pizza()
p1.add_topping('치즈')
p1.add_topping('베이컨')
p1.add_topping('고구마')
p2 = Pizza()
p2.add_topping('소시지')
p2.add_topping('불고기')
p2.add_topping('피망')
p2.remove('소시지')
print(p1.show_topping())
print(p2.show_topping())
```

```python
class Product:
    def __init__ (self, name, code, count):
        self.name = name
        self.code = code
        self.count = count
        
    def show_info(self):
        return 'name=' + self.name + ', code=' + self.code + ', count=' + str(self.count)


p1 = Product('clock', 'c0001', 500)
print(p1.show_info())
p2 = Product('tv', 't0001', 1500)
print(p2.show_info())
```

 ### 각 자리 수의 합을 구하는 EachSum 클래스의 compute() 메소드 작성

```python
class EachSum:
    def __init__(self):
        self.total = 0

 # 코드 작성
    def compute(self,num):
        result = sum([int(i) for i in str(num)])
        return result
each = EachSum()
total = each.compute(12345)
print('합계', total)
```

### 생성자로 메뉴 리스트를 넘겨받은 후 get_menu() 메소드를 통해 메뉴를 선택해주는 SelectMenu 클래스 작성 (random 기능 이용)

```python
import random
class SelectMenu:
    def __init__(self, menu_ls):
        self.menu_ls = menu_ls
        
    def get_menu(self):
        choice = random.choice(self.menu_ls)
        return choice
    
    def get_menu2(self):
        choice = self.menu_ls[random.randint(0,len(self.menu_ls)-1)]
        return choice
    
 # 코드 작성
menu = SelectMenu(['짬뽕', '초밥', '쌀국수', '주꾸미'])
print(menu.get_menu())
print(menu.get_menu2())
```

### 사용된 단어의 횟수 알아내기

[ex]1-document.txt 파일은 영어 단어로 구성된 일반 텍스트 문서이다.

단어는 공백으로 구분된 문자열(특수 기호 포함) 이라고 정의한다.

단어와 사용된 횟수와 함께 출력.

단, 대소문자가 다른 경우 다른 단어라고 정의한다.

```python
from collections import Counter

with open('txt_files/[ex]1-document.txt', 'rt') as file:
    datas = file.read().split() #리스트 형식으로 모두 가져온 후 

word_list = []
for data in datas:
    word_list.append(data)

result=Counter(word_list)
print(result)
```

```python
with open('txt_files/[ex]1-document.txt', 'rt') as file:
    datas = file.read().split() #리스트 형식으로 모두 가져온 후 

result= {}

for i in datas:
    result[i]= datas.count(i)
print(result)
```

```python
with open('txt_files/[ex]1-document.txt', 'rt') as file:
    text = file.read()
text = text.replace('\n', ' ')
words = text.split(' ')
word_dict = dict()

for word in words:
    if word in word_dict:
        word_dict[word] = word_dict[word] +1
    else:
        word_dict[word] = 1
print(word_dict)
    
```

```python
import operator

word_list=sorted(word_dict.items(), key=operator.itemgetter(1), reverse=True) #딕셔너리정렬
                                    #itemgetter(1) => 요소가 2개있는데 그 중 1번째(2번째) 요소를 기준으로 sort, reverse = 내림차순
word_list[:5]
```
