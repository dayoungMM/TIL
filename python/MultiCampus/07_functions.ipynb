 그래프와 같이 이동한 거리 dx, dy 를 구할 수 있는 distance() 함수 작성

```python
def distance(start, end):
    ax, ay = start
    bx, by = end
    return abs(ax-bx), abs(ay-by)
start = (1, 1)
end = (5, 3)
dx, dy = distance(start, end)
print('dx:', dx, 'dy:', dy)
```

```python
string1 = 'ASCII TABLE'
string2 = 'PYTHON'
def show_ch_to_ascii(string):
    temp = ''.join(string.split())
    for i in temp:
        print(ord(i), end=' ')
    print()
show_ch_to_ascii(string1)
show_ch_to_ascii(string2)
```

### pickle
- 객체의 형태를 그대로 유지하면서 파일에 저장하고 불러올 수 있게 하는 모듈
- pickle.dump() : 저장하기
- pickle.load() : 불러오기

```python
import pickle
class NowData:
    def __init__(self):
        self.x = 0
        self.y = 0
    def move(self, x, y):
        self.x = x
        self.y = y
    def get_location(self):
        return self.x, self.y

data = NowData()
data.move(100, 200)

file = open('txt_files/nowdata.bin', 'wb')
pickle.dump(data,file)
file.close()
```

```python
import pickle
file = open('txt_files/nowdata.bin','rb')
data = pickle.load(file)
# data = NowData()
data.get_location()
file.close()
```
