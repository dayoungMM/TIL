## 복사

```python
#1. 얕은 복사(원본과 연결. 함부로 수정하지 말것)
list1 = [1,2,3]
list2 = list1
list2[0] = 100
list1
```

```python
# 2.깊은복사(맘대로 수정가능)
list3 = list1[:]
list3[0]=200
list1
```

```python
# 2.2 깊은복사 방법 2
list3 = list1.copy()
list3[0]=200
list1 
```
