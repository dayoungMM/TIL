### 연습문제 1
파일의 내용을 읽어오기 위한 아래의 코드를
예외처리 구문을 이용하여 결과와 같이 나올 수 있도록 수정

```python
try:
    file = open('text.txt', 'r')
except:
    print("예외발생 [Errno 2] No such file or directory: 'text.txt'")
else:
    text = file.read()
    print(text)
```

```python
try:
    filename = input()
    file = open('txt_files/{}'.format(filename), 'r')
except:
    print("{} 파일이 존재하지 않음".format(filename))
else:
    text = file.read()
    print(text[:10])
finally:
    file.close()
```
