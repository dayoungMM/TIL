### Quiz 1. 어떤 성별을 가진 말이 성적이 좋을까?

- race 데이터 프레임 : 성적데이터
- horse 데이터 프레임 : 성별 데이터
- race 데이터와 horse 데이터를 ID를 key값으로 merge

- 분석 절차
    - 1. 성별에 따른 순위 데이터 프레임으로 생성
    - 2. 전체 데이터에서 성별 데이터를 나눠서 확률 데이터로 변경
    - 3. 1 ~ 5등까지의 성별에 따른 등수 확률 데이터를 출력

```python
df_merged = pd.merge(df_race, df_horse)
df_merged.tail(3)
```

```python
# 1. 성별에 따른 순위 데이터 프레임으로 생성
rank_sex=df_merged.groupby(["SEX","RANK"]).size().reset_index(name="counts")
rank_sex=rank_sex.pivot("RANK","SEX","counts")
rank_sex.head()
```

```python
# 성별에 대한 전체 데이터
df_sex=df_merged.groupby(["SEX"]).size()
df_sex.index[0]
```

```python
# 2. 전체 데이터에서 성별 데이터를 나눠서 확률 데이터로 변경
for i in range(3):
    rank_sex[df_sex.index[i]]=rank_sex[df_sex.index[i]].apply(lambda x: (x/df_sex[i])*100)
rank_sex.head()
```

```python
# 소수 둘째자리까지만 출력
# 결론 : 수컷이 좋은 성적을 낼 확률이 높다.
rank_sex = round(rank_sex, 2)
rank_sex.head()
```

###  Quiz 2. 순위를 승점제로 했을때 2014년 시즌에서 말들의 성적은?

- 1등 : 3점
- 2등 : 2점
- 3등 : 1점

```python
def make_point(x):
    if x==1.0:
        return 3
    elif x==2.0:
        return 2
    elif x==3.0:
        return 1
```

```python
df=df_merged[["RCHOSE_NM","RANK"]]
#len(df["RCHOSE_NM"].unique())==len(df["RCHOSE_NM"]) 중복된 이름이 있다
df['POINT']=df["RANK"].apply(make_point)
df.fillna(value=0, inplace=True)
df
```

```python
df=df.pivot_table("POINT",["RCHOSE_NM"], aggfunc=np.sum)
```

```python
df.sort_values(["POINT"], ascending= False, inplace=True)
df.reset_index().head()
```
