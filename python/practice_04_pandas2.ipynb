### quiz
- makedata모듈을 이용해서 데이터 프레임 만들기
- user_df
    - 8명의 데이터가 있는 데이터 프레임을 만드세요
    - UserID: 1~8
    - Name: makedata.get_name()
    - Age: makedata.get_age()
    - 중복되는 Name값이 없도록

```python
#방법1: 열로 추가하기
datas={}
datas['UserID']=list(range(1,9))
datas['Age']=[makedata.get_age() for _ in range(8)]
datas

names=[]
while True:
    name=makedata.get_name()
    if name not in names:
        names.append(name)
    if len(names)==8:
        break
datas['Name']=names

#datas
df=pd.DataFrame(datas)
df
```

```python
# 방법2: 행별로 추가하기
user_df= pd.DataFrame(columns=['UserID','Age','Name'])


for idx in range(1,9):
    name= makedata.get_name()
    while name in list(user_df["Name"]):
        name= makedata.get_name()
    data={"Name":name, 'UserID':idx, 'Age': makedata.get_age()}
    
    user_df.loc[len(user_df)]=data
user_df
```

```python
#load

titanic=pd.read_csv("datas/train.csv")
```

```python
# 타이타닉 df에서 Fare칼럼 변형하기
titanic_df= titanic[["Survived","Fare"]].copy()
titanic_df.tail()
```

```python
titanic_df['Fare_10']=titanic_df['Fare'].apply(lambda x: x-x%10)
titanic_df.tail()
```

```python
#groupby를 이용하여 counts 만들기
titanic_df_gb= titanic_df.groupby(["Survived","Fare_10"]).size().reset_index(name="counts")
titanic_df_gb.tail()
```

```python
#index가 Fare_10, Coulumn이 Survived, value가 counts인 데이터 프레임 만들기
titanic_df_pv=titanic_df_gb.pivot("Fare_10","Survived","counts")
titanic_df_pv.tail()
```

```python
titanic_pvt=titanic_df_gb.pivot_table("counts",["Fare_10", "Survived"], aggfunc=np.sum)
titanic_pvt.tail()
```
